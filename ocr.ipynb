```python
import easyocr
import pandas as pd
reader = easyocr.Reader(['en', 'es'])
```

```python
def read_text(img):
    result = reader.readtext(f'meme/{img}')
    res = ""
    for (bbox, text, prob) in result:
        if prob > 0.7:
            res += f" {text}"
    res.strip()
    return res
```

```python
df = pd.read_csv("dataset.csv")
```

```python
df['text_in_image'] = df.img.apply(read_text)
```
